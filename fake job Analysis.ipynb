```python
f["department"] = f["department"].astype(str)
f["requirements"] = f["requirements"].astype(str)
f["description"] = f["description"].astype(str)
f["company_profile"] = f["company_profile"].astype(str)
f["benefits"] = f["benefits"].astype(str)
f["required_education"] = f["required_education"].astype(str)
f["company_profile"] = f["company_profile"].astype(str)
f["telecommuting"] = f["telecommuting"].astype(str)
f["has_company_logo"] = f["has_company_logo"].astype(str)
f["has_questions"] = f["has_questions"].astype(str)
```

```python
f["X"] = f["department"]+ "  " + f["requirements"]+ "  " + f["description"]+"  " + f["company_profile"]+ "  " + f["benefits"]+ "  " + f["required_education"]
```

```python
f=f.loc[:,['fraudulent','X']]
f.head(2)
```

```python
f.rename(columns={'fraudulent': 'Y'},inplace = True)
f.head(2)
```

```python
f['X'][0]
```

```python
f['X']=f['X'].str.lower()
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
l1=list(stopwords.words('english'))
l1
```

```python
import string 
string.punctuation
```

```python
def text_process(mess):            ### creating a function

    """                                                        ## a docstring

    1. remove the punctuation

    2. remove the stopwords

    3. return the list of clean textwords

    

    """

    nopunc = [char for char in mess if char not in string.punctuation]

    nopunc = "".join(nopunc)

    

    return [ word for word in nopunc.split() if word not in l1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_transformer   = CountVectorizer(analyzer = text_process).fit(f['X'])

bow_transformer.vocabulary_  #counting the words multiple times
```

```python
len(bow_transformer.vocabulary_ )
```

```python
f_bow = bow_transformer.transform(f.X)
```

```python
f_bow.shape
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x , train_y,test_y = train_test_split(f_bow,f.Y,random_state = 123 , test_size = 0.2)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
from sklearn.linear_model import LogisticRegression
lr_f = LogisticRegression(class_weight='balanced')
lr_f.fit(train_x,train_y)
pred_f = lr_f.predict(test_x)
tab_f = confusion_matrix(test_y,pred_f)
print(tab_f)
print('accuracy is' , tab_f.diagonal().sum()* 100 / tab_f.sum())
print("--------------------------")
print(classification_report(test_y,pred_f))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt= DecisionTreeClassifier()
dt.fit(train_x,train_y)

pred_dt = dt.predict(test_x)
pred_dt

confusion_matrix(test_y,pred_dt)

tab_dt = confusion_matrix(test_y,pred_dt)
print(tab_dt)
print('accuracy is',tab_dt.diagonal().sum()*100/tab_dt.sum())
print("--------------------------")
print(classification_report(test_y,pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc= RandomForestClassifier()
rfc.fit(train_x,train_y)

pred_rfc = rfc.predict(test_x)
pred_rfc

confusion_matrix(test_y,pred_rfc)

tab_rfc = confusion_matrix(test_y,pred_rfc)
print(tab_rfc)
print('accuracy is',tab_rfc.diagonal().sum()*100/tab_rfc.sum())
print("--------------------------")
print(classification_report(test_y,pred_rfc))
```

```python
from sklearn.svm import SVC
svc=SVC(class_weight='balanced')
svc.fit(train_x,train_y)
pred_svc = svc.predict(test_x)
print(confusion_matrix(test_y,pred_svc))
print("--------------------------")
print(classification_report(test_y,pred_svc))
```
